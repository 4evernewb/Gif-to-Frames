```python
#Without saving the frames but to dispay only.
from PIL import Image, ImageSequence

im = Image.open("congo.gif")

index = 1
for frame in ImageSequence.Iterator(im):
    display(frame)
    frame.show()
    index += 1
```

```python
#To display and save it in folder named frames
from PIL import Image, ImageSequence

im = Image.open("congo.gif")

index = 1
for frame in ImageSequence.Iterator(im):
    frame.save("frames/frame%d.png" % index)
    display(frame)
    frame.show()
    index += 1
```
